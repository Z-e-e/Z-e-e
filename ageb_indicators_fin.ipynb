```
#df1 reduced df
columns = ['ENTIDAD', 'MUN', 'LOC', 'AGEB', 'MZA','VPH_REFRI', 'VPH_LAVAD','VPH_TV', 'VPH_PC','GRAPROES', 'P18YM_PB', 'P_18YMAS']

df1 = pd.DataFrame()

for col in columns:
  df1[col] = df[col]


df1.head()

#df2 copy
df2 = df1.copy()

#new df2
df2['AGEB_Key']=df.apply(lambda x:'%s%s%s%s' % (x['ENTIDAD'],x['MUN'],x['LOC'],x['AGEB']),axis=1)
df2.head()

#ageb1 copy
ageb1 = df2.copy()
#obj to num
columns = ['VPH_REFRI', 'VPH_LAVAD','VPH_TV', 'VPH_PC','GRAPROES', 'P18YM_PB', 'P_18YMAS']

for col in columns:
  ageb1[col] = pd.to_numeric(ageb1[col], errors = 'coerce')

ageb1.head()

#pop with refrigirators, washers, TVs and PCs
ageb1['pop_4_products'] = ageb1['VPH_REFRI'] + ageb1['VPH_LAVAD'] + ageb1['VPH_TV'] + ageb1['VPH_PC'] #double counting posible

ageb1['compu'] = ageb1['VPH_PC'] / ageb1['pop_4_products']

#average level of education (132)
ageb1['esc'] = ageb1['GRAPROES']/ageb1['GRAPROES'].max()

#population aged 18+ w/ post-basic education / population of people 18 yo and older
ageb1['edusup'] = ageb1['P18YM_PB'] / ageb1['P_18YMAS']

ageb1.head()
```

```
imp_cols = ['compu','esc', 'edusup']
ageb1 = ageb1.dropna(subset = imp_cols)

ind_trial = ageb1.iloc[:, -3:]
ind_trial.head()

print(ind_trial.corr())

#PCA
indicators = ['compu', 'esc', 'edusup' ]
x = ageb1.loc[:, indicators].values

x = StandardScaler().fit_transform(x)

print(np.mean(x),np.std(x))

print(x[:5])

print(range(x.shape[1]))

print(x.shape[1])

print(x[:5,0])

ind_cols = ['comp', 'esc', 'edusup' ]
#for i in range(x.shape[1]):
#  'indicators'+str(i)
norm_ind = pd.DataFrame(x,columns=ind_cols)
norm_ind.head()


inds = ageb1.iloc[:, -3:]
#inds.head()
print(norm_ind.corr())

pca_ind = PCA(n_components=3)
pc_ind = pca_ind.fit_transform(x)

print(pc_ind)

pc_df = pd.DataFrame(data = pc_ind
             , columns = ['pc_1', 'pc_2', 'pc_3'])

print(pc_df.head())

print("pc_df length".format(len(pc_df)))

print('variance explained per component: {}'.format(pca_ind.explained_variance_ratio_))


ageb_1 = ageb1.reset_index()
print(ageb_1.head())

pca_ageb = pd.merge(ageb_1, pc_df, left_index = True, right_index = True)
print(pca_ageb.head())

pca_ageb['rating'] = np.where(pca_ageb['pc_1']<=-0.4, 'E', 
                              (np.where(pca_ageb['pc_1']<=0.05, 'D',
                                        (np.where(pca_ageb['pc_1']<=0.34, 'D+',
                                                   (np.where(pca_ageb['pc_1']<=0.68, 'C-',
                                                              (np.where(pca_ageb['pc_1']<=1.25, 'C',
                                                                         (np.where(pca_ageb['pc_1']<=2.41, 'C+',
                                                                                    (np.where(pca_ageb['pc_1']>2.41, 'A/B',0)))))))))))))


pca_ageb1 = pca_ageb.copy()
pca_ageb1.head()
```
